```python
#chi squared before imputing
#chi squared test for feature and target
#need the frequencies for the values
from scipy.stats import chi2_contingency

def chi(data, target):

    for c in data.columns:
        print(c)
        data_crosstab = pd.crosstab(data[c], 
                                    target,  
                                       margins = False) 
        stat, p, dof, expected = chi2_contingency(data_crosstab)
        # select significance value

        alpha = 0.05
        # Determine whether to reject or keep your null hypothesis
        print(alpha, p)
        if p <= alpha:
            print('Variables are associated (reject H0)')
        else:
            print('Variables are not associated(fail to reject H0)')
```

```python
chi(consolidated_data, consolidated_data['SHTFLU12M_A'])
```

```python
#impute values with 2 for the questions that were only asked to people who answered "yes" to other questions
#make list of columns that make sense to do this
impute = ['HYPMED_A', 'CHLMED_A', 'ASTILL_A', 'ASAT12M_A', 'ASER12M_A', 'DIBEV_A']
for col in impute:
    mask = pd.isnull(consolidated_data[col])
    consolidated_data[col][mask == True] = 2
```

```python
#want to check the percentage of data of each feature that is not 1 or 2 (these are the major answers for each question--yes/no)
p = .1
newcols = pd.DataFrame()
for i in range(len(consolidated_data.columns)):
    col = consolidated_data.columns[i]
    index_vals = list(consolidated_data[col].value_counts().index)
    vals = [consolidated_data[col].value_counts()[index_vals[j]] for j in range(len(index_vals)) if index_vals[j] != 1 and index_vals[j] != 2]
    ind = [index_vals[j] for j in range(len(index_vals)) if index_vals[j] != 1 and index_vals[j] != 2]
    percentage = sum(vals)/len(consolidated_data)
    print(ind)
    #Binning non yes/no values if the percentage is less than 10
    if percentage < p:
        index = []
        for k in range(len(ind)):
            index.append(list(np.where(consolidated_data[col] == ind[k])))
            t = pd.DataFrame(index)
            consolidated_data[col].loc[t[0][k]] = 10
          
```
