# Logistic Regression Learning Curve

```python
import numpy as np
train_sizes = np.linspace(1, len(data), 6)/len(data)
```

```python
train_sizes
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import learning_curve

features = data.drop('target', 1)
features
```

```python
target = data.target
```

```python
train_sizes, train_scores, validation_scores = learning_curve(estimator = LogisticRegression(C=0.46415888336127775, class_weight=None, dual=False,
                   fit_intercept=True, intercept_scaling=1, l1_ratio=None,
                   max_iter=100, multi_class='auto', n_jobs=None, penalty='l1',
                   random_state=50, solver='liblinear', tol=0.0001, verbose=0,
                   warm_start=False), X = features, y = data.target, train_sizes = train_sizes, cv = 5, scoring = 'roc_auc')
```

```python
import sklearn
sorted(sklearn.metrics.SCORERS.keys())
```

```python
train_sizes2, train_scores2, validation_scores2 = learning_curve(estimator = LogisticRegression(C=0.46415888336127775, class_weight=None, dual=False,
                   fit_intercept=True, intercept_scaling=1, l1_ratio=None,
                   max_iter=100, multi_class='auto', n_jobs=None, penalty='l1',
                   random_state=50, solver='liblinear', tol=0.0001, verbose=0,
                   warm_start=False), X = features, y = data.target, train_sizes = train_sizes, cv = 5, scoring = 'f1')
```

```python
train_scores_mean = train_scores.mean(axis=1)
validation_scores_mean = validation_scores.mean(axis = 1)
```

```python
import matplotlib.pyplot as plt
plt.plot(train_sizes, train_scores_mean, label = 'Training')
plt.plot(train_sizes, validation_scores_mean, label = 'Validation')
plt.xlabel('Sample size')
plt.ylabel('AUC')
plt.legend()
```

```python
train_scores_mean2 = train_scores2.mean(axis=1)
validation_scores_mean2 = validation_scores2.mean(axis = 1)
plt.plot(train_sizes2, train_scores_mean, label = 'Training')
plt.plot(train_sizes2, validation_scores_mean, label = 'Validation')
plt.xlabel('Sample size')
plt.ylabel('F1')
plt.legend()
```

```python
train_sizes3, train_scores3, validation_scores3 = learning_curve(estimator = LogisticRegression(C=0.46415888336127775, class_weight=None, dual=False,
                   fit_intercept=True, intercept_scaling=1, l1_ratio=None,
                   max_iter=100, multi_class='auto', n_jobs=None, penalty='l1',
                   random_state=50, solver='liblinear', tol=0.0001, verbose=0,
                   warm_start=False), X = features, y = data.target, train_sizes = train_sizes, cv = 5, scoring = 'neg_log_loss')
```

```python
train_scores_mean3 = train_scores3.mean(axis=1)
validation_scores_mean3 = validation_scores3.mean(axis = 1)
plt.plot(train_sizes3, -train_scores_mean3, label = 'Training')
plt.plot(train_sizes3, -validation_scores_mean3, label = 'Validation')
plt.xlabel('Sample size')
plt.ylabel('Log loss')
plt.legend()
```

# Random Forest Learning Curve

```python
from sklearn.ensemble import RandomForestClassifier
train_sizes, train_scores, validation_scores = learning_curve(estimator = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=12,
                       max_features=0.7999999999999999, max_leaf_nodes=33,
                       max_samples=None, min_impurity_decrease=0.0,
                       min_impurity_split=None, min_samples_leaf=1,
                       min_samples_split=2, min_weight_fraction_leaf=0.0,
                       n_estimators=188, n_jobs=None, oob_score=True,
                       random_state=50, verbose=0, warm_start=False), X = features, y = data.target, train_sizes = train_sizes, cv = 5, scoring = 'accuracy')
```

```python
train_mean = np.mean(train_scores, 1)
validation_mean = np.mean(validation_scores, 1)
```

```python
plt.plot(train_sizes, train_mean, label = 'Training')
plt.plot(train_sizes, validation_mean, label = 'Validation')
plt.xlabel('Sample Size')
plt.ylabel('Accuracy')
plt.legend()
```
